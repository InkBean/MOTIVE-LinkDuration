```python
import tensorflow as tf
from tensorflow import keras

from time import time
from tensorflow.python.keras.callbacks import TensorBoard

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
PATH = os.getcwd()

X_train = pd.read_csv(PATH + "/Train/X_train.csv", header=0)
Y_train = pd.read_csv(PATH + "/Train/Y_train.csv", header=0)

X_test = pd.read_csv(PATH + "/Test/X_test.csv", header=0)
Y_test = pd.read_csv(PATH + "/Test/Y_test.csv", header=0)
```

```python
tensorboard = TensorBoard(log_dir = "logs/{}".format(time()))
```

```python
inputs = keras.Input(shape=(73,))
h1 = keras.layers.Dense(256, activation='relu')(inputs)
h1 = keras.layers.Dropout(0.5)(h1)
#block1_out = layers.MaxPooling2D(3)(h1)

h2 = keras.layers.Dense(256, activation='relu')(h1)
h2 = keras.layers.Dropout(0.5)(h2)
block2_out = keras.layers.add([h2, h1])

h3 = keras.layers.Dense(256, activation='relu')(block2_out)
h3 = keras.layers.Dropout(0.5)(h3)

h4 = keras.layers.Dense(256, activation='relu')(h3)
h4 = keras.layers.Dropout(0.5)(h4)

block4_out = keras.layers.add([h4, h3])

outputs = keras.layers.Dense(3, activation='softmax')(block4_out)

model = keras.Model(inputs, outputs, name='small resnet')
model.summary()

adm = keras.optimizers.Adam(lr=0.00005, beta_1=0.9, beta_2=0.999, epsilon=None,  amsgrad=False)
#Nadam = keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)

model.compile(optimizer= adm,
              loss='categorical_crossentropy',
              metrics=['accuracy']) 

#model.fit(X_train, Y_train, batch_size=64, epochs=1000, shuffle=True, verbose=1, validation_split=0.1)
model.fit(X_train, Y_train, batch_size=64, epochs=400, shuffle=True, verbose=1, validation_split=0.1,callbacks=[tensorboard])
print('\ntest set:')
loss = model.evaluate(X_test, Y_test, batch_size=64, verbose=1)
```

```python
model.save(os.getcwd() + '/NN_models/5layers_residual2.h5')
```
