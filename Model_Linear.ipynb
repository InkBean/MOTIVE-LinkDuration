```python
import tensorflow as tf
from tensorflow import keras

from time import time
from tensorflow.python.keras.callbacks import TensorBoard

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

print(tf.__version__)
```

```python
PATH = os.getcwd()
PATH = PATH + "/data"

df = pd.read_csv(PATH + "/lank_v6.csv", header=0)
data = df.values
```

```python
num_samples = df.shape[0]
num_features = df.shape[1]
print("number of samples = " + str(num_samples))
print("number of features = " + str(num_features))
print(df.columns)
```

```python
#df = df.drop(['Unnamed: 0',
 #             'space_headway_X', 'space_headway_Y'], axis = 1)
data = df.values
num_samples = df.shape[0]
num_features = df.shape[1]
print("number of samples = " + str(num_samples))
print("number of features = " + str(num_features))
print(df.columns)
print(data[0][:])
```

```python
train = data[:len(data) * 9 // 10]
#print(train.shape[1])
X_train = np.array([[row[i] for i in range(0, len(row)-1) ] for row in train])

Y_train = np.array([[x[num_features-1]]for x in train])  # train[:,:1]

test = data[len(data) * 9 // 10:]
X_test = np.array([[row[i] for i in range(0, len(row)-1) ] for row in test])
Y_test = np.array([[x[num_features-1]] for x in test])  # train[:,:1]


print(X_train.shape)
print(Y_train.shape)
print()
print(X_test.shape)
print(Y_test.shape)
print(X_train[0][:])
print(Y_train[0])
print(X_test[0][:])
print(Y_test[0])
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)

Y_train = Y_train / 1000
print(X_train[0][:] )
print(Y_train[0])
```

```python
scaler = StandardScaler()
scaler.fit(X_test)
X_test = scaler.transform(X_test)

Y_test = Y_test / 1000
print(X_test[0][:] )
print(Y_test[0])
```

```python
from tensorflow.compat.v1.keras.utils import to_categorical

Y_train = tf.keras.utils.to_categorical(
    Y_train,
    num_classes=2,
    dtype='float32'
)

Y_test = tf.keras.utils.to_categorical(
    Y_test,
    num_classes=2,
    dtype='float32'
)
```

```python
print(Y_train[0:10])
print(Y_test[0:10])
```

```python
tensorboard = TensorBoard(log_dir = "logs/{}".format(time()))
#, callbacks=[tensorboard]
```

```python
model = keras.Sequential(
)

model.add(keras.layers.Dense(256,input_shape=(36,)))

model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(256))
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(1, activation='linear'))

adm = keras.optimizers.Adam(lr=0.00005, beta_1=0.9, beta_2=0.999, epsilon=None,  amsgrad=False)
#Nadam = keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)

model.compile(optimizer= "adam",
              loss='mae',
              metrics=['mae']) 

#model.fit(X_train, Y_train, batch_size=64, epochs=1000, shuffle=True, verbose=1, validation_split=0.1)
model.fit(X_train, Y_train, batch_size=64, epochs=500, shuffle=True, verbose=1, validation_split=0.1,callbacks=[tensorboard])
print('\ntest set:')
loss = model.evaluate(X_test, Y_test, batch_size=64, verbose=1)
print(" accuracy: %f" % (loss[0]))
```
