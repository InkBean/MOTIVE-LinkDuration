```python
df = pd.read_csv(PATH + "/t1.csv")
after = []

num_samples = df.shape[0]
num_features = df.shape[1]

print(df.columns)
data = df.values 
print(data[1][:])

for i in range(num_samples-1):
    l_duration = data[i][0]
    p_duration = data[i][2]
    
    l = 0
    p = 0
    
    if l_duration >= 10:
        if l_duration >=30:
            l = 3
        else:
            l = 2
    else:
        l = 1
    
    if p_duration >= 10:
        if p_duration >=30:
            p = 3
        else:
            p = 2
    else:
        p = 1
    
    match = 0
    if p == l:
        match = 1
    
    a = list(data[i][:num_features])
    a.extend([match])
    after.append(a)
    
#convert the list to a numpy array        
output_array = np.asarray(after)    

# convert the numpy array to a csv file.
t2 = np.savetxt("t2.csv", output_array, delimiter=",",fmt = "%s")
```
