```python
import tensorflow as tf
from tensorflow import keras

from time import time
from tensorflow.python.keras.callbacks import TensorBoard

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

print(tf.__version__)
```

```python
df = pd.read_csv("new1.csv", header=0)
data = df.values
```

```python
num_samples = df.shape[0]
num_features = df.shape[1]
print("number of samples = " + str(num_samples))
print("number of features = " + str(num_features))
print(df.columns)
```

```python
X = df.drop(['zero_to_10', 'ten_to_30', 'thirty_plus','duration'], axis=1)
#Y = df[['zero_to_10', 'ten_to_30', 'thirty_plus']]
Y = df[['zero_to_10', 'ten_to_30', 'thirty_plus']]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

```python
#from tensorflow.compat.v1.keras.utils import to_categorical

#Y_train = tf.reshape(Y_train,(17778,3))
#Y_test = tf.reshape(Y_test, (1976,3))
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)

#Y_train = Y_train / 1000
print(X_train[0][:] )
#print(Y_train[0])

scaler = StandardScaler()
scaler.fit(X_test)
X_test = scaler.transform(X_test)

#Y_test = Y_test / 1000
print(X_test[0][:] )
#print(Y_test[0])
```

```python
'''
from tensorflow.compat.v1.keras.utils import to_categorical

Y_train = tf.keras.utils.to_categorical(
    Y_train,
    num_classes=None,
    dtype='float32'
)

Y_test = tf.keras.utils.to_categorical(
    Y_test,
    num_classes=None,
    dtype='float32'
)
'''
print(Y_train[0])
print(Y_test[0])
```

```python
tensorboard = TensorBoard(log_dir = "logs/{}".format(time()))
```

```python
model = keras.Sequential(
)

model.add(keras.layers.Dense(256,input_shape=(72,)))

model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(256))
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(3, activation='softmax'))

adm = keras.optimizers.Adam(lr=0.00001, beta_1=0.9, beta_2=0.999, epsilon=None,  amsgrad=False)
#Nadam = keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)

model.compile(optimizer= adm,
              loss='categorical_crossentropy',
              metrics=['accuracy']) 

#model.fit(X_train, Y_train, batch_size=64, epochs=200, shuffle=True, verbose=1, validation_split=0.1)
model.fit(X_train, Y_train, batch_size=64, epochs=5000, shuffle=True, verbose=1, validation_split=0.1,callbacks=[tensorboard])
print('\ntest set:')
loss = model.evaluate(X_test, Y_test, batch_size=128, verbose=1)
#print(" accuracy: %f" % (loss[0]))
```

```python
prediction = model.predict(X_test)
```
